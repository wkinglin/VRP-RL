```python
import torch
import torch.nn as nn
import torch.optim as optim
import time

# --- 1. å®šä¹‰ä¸€ä¸ªç®€å•çš„ç¥ç»ç½‘ç»œ ---
# è¿™æ˜¯ä¸€ä¸ªç®€å•çš„å¤šå±‚æ„ŸçŸ¥æœºï¼ˆMLPï¼‰
class SimpleNet(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(SimpleNet, self).__init__()
        self.layer1 = nn.Linear(input_size, hidden_size)
        self.relu = nn.ReLU()
        self.layer2 = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        out = self.layer1(x)
        out = self.relu(out)
        out = self.layer2(out)
        return out
```

```python
# --- 2. æ£€æŸ¥å¹¶è®¾ç½®è®¾å¤‡ (GPUæˆ–CPU) ---
# è¿™æ˜¯å…³é”®æ­¥éª¤ï¼šæ£€æŸ¥torch.cudaæ˜¯å¦å¯ç”¨
print(torch.__version__)
if torch.cuda.is_available():
    device = torch.device("cuda")
    print(f"ğŸ‰ æˆåŠŸï¼æ£€æµ‹åˆ°å¹¶æ­£åœ¨ä½¿ç”¨è®¾å¤‡: {torch.cuda.get_device_name(0)}")
else:
    device = torch.device("cpu")
    print("âš ï¸ æœªæ£€æµ‹åˆ°CUDA GPUï¼Œæ­£åœ¨ä½¿ç”¨è®¾å¤‡: CPU")

# --- 3. è®¾ç½®è¶…å‚æ•° ---
input_size = 512    # è¾“å…¥ç‰¹å¾ç»´åº¦
hidden_size = 1024  # éšè—å±‚å¤§å°
output_size = 10    # è¾“å‡ºç±»åˆ«æ•°
num_epochs = 100    # è®­ç»ƒè½®æ¬¡
learning_rate = 0.001 # å­¦ä¹ ç‡
batch_size = 64     # æ¯æ‰¹æ¬¡å¤§å°
num_samples = 10000 # æ ·æœ¬æ€»æ•°

# --- 4. å®ä¾‹åŒ–æ¨¡å‹ã€æŸå¤±å‡½æ•°å’Œä¼˜åŒ–å™¨ ---
model = SimpleNet(input_size, hidden_size, output_size)

# å°†æ¨¡å‹ç§»åŠ¨åˆ°é€‰å®šçš„è®¾å¤‡ (GPUæˆ–CPU)
model.to(device)

# å®šä¹‰æŸå¤±å‡½æ•° (ä¾‹å¦‚ï¼šå‡æ–¹è¯¯å·®æŸå¤±)
criterion = nn.MSELoss()

# å®šä¹‰ä¼˜åŒ–å™¨ (ä¾‹å¦‚ï¼šAdam)
optimizer = optim.Adam(model.parameters(), lr=learning_rate)

print("\næ¨¡å‹å·²æˆåŠŸç§»åŠ¨åˆ°:", next(model.parameters()).device)

# --- 5. åˆ›å»ºæ¨¡æ‹Ÿæ•°æ® ---
# åˆ›å»ºéšæœºçš„è¾“å…¥æ•°æ®å’Œç›®æ ‡æ•°æ®
# .to(device) ä¼šå°†è¿™äº›æ•°æ®å¼ é‡ä¹Ÿç§»åŠ¨åˆ°GPU/CPU
X_train = torch.randn(num_samples, input_size).to(device)
y_train = torch.randn(num_samples, output_size).to(device)

print(f"æ•°æ®å·²åˆ›å»ºï¼Œå°†åœ¨ {device} ä¸Šè¿›è¡Œè®­ç»ƒ...")
print("-" * 30)

# --- 6. å¼€å§‹è®­ç»ƒå¹¶è®¡æ—¶ ---
start_time = time.time()

for epoch in range(num_epochs):
    # åœ¨è¿™ä¸ªç®€å•çš„ä¾‹å­ä¸­ï¼Œæˆ‘ä»¬ä¸€æ¬¡æ€§å¤„ç†æ‰€æœ‰æ•°æ®
    # å®é™…é¡¹ç›®ä¸­ä½ ä¼šä½¿ç”¨DataLoaderæ¥åˆ†æ‰¹å¤„ç†
    
    # å‰å‘ä¼ æ’­
    outputs = model(X_train)
    loss = criterion(outputs, y_train)
    
    # åå‘ä¼ æ’­å’Œä¼˜åŒ–
    optimizer.zero_grad()  # æ¸…ç©ºä¹‹å‰çš„æ¢¯åº¦
    loss.backward()        # è®¡ç®—å½“å‰æ¢¯åº¦
    optimizer.step()       # æ›´æ–°æƒé‡
    
    if (epoch + 1) % 10 == 0:
        print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}')

end_time = time.time()

# --- 7. æ‰“å°ç»“æœ ---
total_time = end_time - start_time
print("-" * 30)
print(f"è®­ç»ƒå®Œæˆ!")
print(f"åœ¨è®¾å¤‡ {device} ä¸Šæ€»å…±ç”¨æ—¶: {total_time:.2f} ç§’")
```
